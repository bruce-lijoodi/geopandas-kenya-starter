```python
!pip install geopandas matplotlib requests

import geopandas as gpd
import matplotlib.pyplot as plt
import requests
import zipfile
import io
import os

# Reliable mirror for Natural Earth data (from GitHub)
url = "https://github.com/nvkelso/natural-earth-vector/archive/refs/heads/master.zip"
r = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall("natural_earth")

# Path to countries shapefile
shapefile_path = "natural_earth/natural-earth-vector-master/10m_cultural/ne_10m_admin_0_countries.shp"

# Read the shapefile
world = gpd.read_file(shapefile_path)

# Filter Kenya
kenya = world[world['NAME'] == 'Kenya']

# Plot
fig, ax = plt.subplots(figsize=(6, 6))
kenya.plot(ax=ax, color='lightgreen', edgecolor='black')

ax.set_title("Map of Kenya", fontsize=14)
plt.show()
```
